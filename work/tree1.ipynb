```python
empleo = ["si", "si", "no", "si", "si", "no"]
ingreso = [3000, 1000, 0, 1500, 2000, 0]
credito = ["aprobado", "rechazado", "rechazado", "aprobado", "aprobado", "rechazado"]
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

empleo_encoded = le.fit_transform(empleo)
credito_encoded = le.fit_transform(credito)

print(empleo_encoded)
print(credito_encoded)
```

```python
features = list(zip(empleo_encoded,ingreso))
features
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy', random_state=0, splitter="random")
tree = tree.fit(features, credito_encoded)
```

Graficar arbol

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plot_tree(tree, filled=True)
plt.show()

#tiene empleos, ingresos
```
