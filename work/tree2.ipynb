```python
a = [
    10.5,
    9.8,
    11.2,
    9.2,
    9.1,
    10.8,
    11.5,
    11.2,
    10.3,
    9.7,
]
b = [
    1,
    5,
    20,
    60,
    90,
    120,
    180,
    220,
    340,
    410,
]
c = [
    400.5,
    405.8,
    410.9,
    425.4,
    450.1,
    500.5,
    575.7,
    456.5,
    320.1,
    221.9,
]

d = [
    "TRUE",
    "FALSE",
    "FALSE",
    "FALSE",
    "TRUE",
    "TRUE",
    "FALSE",
    "TRUE",
    "TRUE",
    "FALSE",
]

y = [
    "N",
    "P",
    "P",
    "N",
    "P",
    "P",
    "N",
    "P",
    "N",
    "P",
]
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

a_encoded = a
b_encoded = b
c_encoded = c
d_encoded = le.fit_transform(d)
y_encoded = le.fit_transform(y)

print("a", a_encoded)
print("b", b_encoded)
print("c", c_encoded)
print("d", d_encoded)
print("y", y_encoded)
```

```python
features = list(zip(a_encoded, b_encoded, c_encoded, d_encoded))
features
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy', random_state=0, splitter="random")
tree = tree.fit(features, y_encoded)
```

Graficar arbol

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plot_tree(tree, filled=True)
plt.show()
```
