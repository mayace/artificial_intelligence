```python
a = [
    "sunny",
    "sunny",
    "overcast",
    "rain",
    "rain",
    "rain",
    "overcast",
    "sunny",
    "sunny",
    "rain",
    "sunny",
    "overcast",
    "overcast",
    "rain",
]
b = [
    "hot",
    "hot",
    "hot",
    "mild",
    "cold",
    "cold",
    "cold",
    "mild",
    "cold",
    "mild",
    "mild",
    "mild",
    "hot",
    "mild",
]
c = [
    "high",
    "high",
    "high",
    "high",
    "normal",
    "normal",
    "normal",
    "high",
    "normal",
    "normal",
    "normal",
    "high",
    "normal",
    "high",
]

d = [
    "false",
    "true",
    "false",
    "false",
    "false",
    "true",
    "true",
    "false",
    "false",
    "false",
    "true",
    "true",
    "false",
    "true",
]

y = [
    "N",
    "N",
    "P",
    "P",
    "P",
    "N",
    "P",
    "N",
    "P",
    "P",
    "P",
    "P",
    "P",
    "N",
]
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

a_encoded = le.fit_transform(a)
b_encoded = le.fit_transform(b)
c_encoded = le.fit_transform(c)
d_encoded = le.fit_transform(d)
y_encoded = le.fit_transform(y)

print("a", a_encoded)
print("b", b_encoded)
print("c", c_encoded)
print("d", d_encoded)
print("y", y_encoded)
```

```python
features = list(zip(a_encoded, b_encoded, c_encoded, d_encoded))
features
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(criterion='entropy', random_state=0, splitter="random")
tree = tree.fit(features, y_encoded)
```

Graficar arbol

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plot_tree(tree, filled=True)
plt.show()

#outlook, temperature, humidity, windy
```
