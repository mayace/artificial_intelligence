```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import numpy as np

w = [80, 60, 60, 40]
o = [0.33, 0.15, 0.15, 0.16]
v = [3.7, 3.7, 3.7, 3.7]
a = [35, 30, 20, 15]
# p = [217,197,240,240]
p = [139,197,176,186]

features=list(zip(w, o, v, a))
clf = DecisionTreeClassifier(criterion='entropy',
                             random_state=0,
                             splitter='random').fit(features, p)
plot_tree(clf, filled=True)
plt.show()


x = np.array([40, 0.6, 3.7, 20]).reshape(1, -1)
print(clf.predict(x))

x = np.array([40, 0.6, 3.7, 20]).reshape(1, -1)
print(clf.predict(x))

x = np.array( [75 ,0.15, 3.7, 25]).reshape(1, -1)
print(clf.predict(x))

x = np.array( [75, 0.15, 3.7, 4000]).reshape(1, -1)
print(clf.predict(x))


x = np.array([60 ,0.2 ,3.7 ,20]).reshape(1, -1)
print(clf.predict(x))
```
