```python
import pandas as pd

df = pd.read_csv('./work/2s2023/pred.csv')
```

```python
df
```

```python
X = df.drop(columns=["NO", "E"])
X
```

```python
y = df["E"]
y
```

```python
features = X
features
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model = model.fit(features, y)
model.get_params()
```

```python
print("original\t", y)
print("bayes\t\t", model.predict(features))
```
